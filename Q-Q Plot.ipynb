```python
# Q-Q Plot
import numpy as np
import pylab
import scipy.stats as stats

# N(0,1) ,HERE loc is mean and scale is std-deviation ,we have genrated random observations from gaussian distribution.

std_normal = np.random.normal(loc=0 , scale = 1, size=1000)

# 0th to 100th percentile of std-normal
for i in range(0,101):
    print(i,(np.percentile(std_normal,i)))
```

```python
# generates 100 samples from N(20,5)
measurments = np.random.normal(loc=20,scale = 5,size=100)
# try size=1000 , then the point will lie in the straight line,as the size is increasing.

# Q-Q plot or Probability plot using probplot as stats.probplot(),"https://stackoverflow.com/questions/48108582/how-to-interpret-scipy-stats-probplot-results"
# dist = "norm" ,means compute the value with N(0,1)
stats.probplot(measurments,dist="norm",plot=pylab)
pylab.show()
```

###### Observation: Here both x and y are gaussian in the plot with y as N(0,1) and x as N(20,5).
###### some points are not lieing in the straight line,and if we change the size from 100 to 1000 than most of the points lies on the straight line,so as the sample size increases than the most of the points lies between the straight line ,So one limitation is if obseervation is small it is difficult to recognise Q-Q plot.


```python
measurments = np.random.uniform(low=-1,high=1,size=1000)
stats.probplot(measurments,dist="norm",plot=pylab)
pylab.show()
```

###### Observation: If the points are diverging more and more from the straight line then the distribution of x axis and y axis are different such as unifor distribution in x axis and gaussian distribution in the y axis.
